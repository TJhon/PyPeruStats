# INGRESOS

```python
# Steps para años 2009-2011
STEPS_INGRESO_2009_2011 = [
    dict(
        desc="Seleccionar total, click nivel de gobierno",
        select_row_text="total",
        click_bttn="nivel_gobierno",
    ),
    dict(
        desc="Seleccionar gob locales, click genérica",
        select_row_text="locales",
        click_bttn="generica",
    ),
    dict(
        name="estructura_generica",
        desc="Iterar estructuras de recaudacion, click departamento",
        select_row_text=None,  # None = loop
        click_bttn="departamento",
        save_group=True,
    ),
    dict(
        name="departamento",
        desc="Iterar departamentos, click municipalidad",
        select_row_text=None,
        click_bttn="municipalidad",
    ),
]

# Steps para años 2012-2026
STEPS_INGRESO_2012_2026 = [
    dict(
        desc="Seleccionar total, click nivel de gobierno",
        select_row_text="total",
        click_bttn="nivel_gobierno",
    ),
    dict(
        desc="Seleccionar gob locales, click sub tipo",
        select_row_text="locales",
        click_bttn="sub_tipo_gobierno",
    ),
    dict(
        desc="Seleccionar gob. municipales, click genérica",
        select_row_text="municipalidades",
        click_bttn="generica",
    ),
    dict(
        name="estructura_generica",
        desc="Iterar estructuras de recaudacion, click departamento",
        select_row_text=None,
        click_bttn="departamento",
        save_group=True,
    ),
    dict(
        name="departamento",
        desc="Iterar departamentos, click municipalidad",
        select_row_text=None,
        click_bttn="municipalidad",
    ),
]
```

```python
scraper_ingresos = MEFScraper(
        tipo='ingreso',
        master_dir_save='./data/raw/temp/ingresos/'
    )


scraper_ingresos.run(year=2009, steps=STEPS_INGRESO_2009_2011)
```

```python
scraper_ingresos = MEFScraper(
        tipo='ingreso',
        master_dir_save='./data/raw/temp/ingresos/'
    )


scraper_ingresos.run(year=2019, steps=STEPS_INGRESO_2012_2026)
```
