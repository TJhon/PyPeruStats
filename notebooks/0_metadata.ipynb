```python
import pandas as pd
url =  'https://estadisticas.bcrp.gob.pe/estadisticas/series/metadata'

metadata = pd.read_csv(url, delimiter=';', encoding='latin-1')

metadata.head(2)
```

```python
# Como esta estructurado los datos con codigo y frecuencia

# Crear una copia explícita de las columnas seleccionadas
freq = metadata[['Código de serie', 'Frecuencia']].copy()

# Renombrar columnas
freq.columns = ['cod', 'freq']

# Asegurarse de que los datos en 'cod' sean cadenas
freq['cod'] = freq['cod'].astype(str)

# Usar .loc para modificar columnas y evitar SettingWithCopyWarning
freq.loc[:, 'last_string'] = freq['cod'].apply(lambda x: x[-1] if len(x) > 0 else "")
freq.loc[:, 'first_string'] = freq['cod'].apply(lambda x: x[0] if len(x) > 0 else "")

# Verificar el resultado
freq.head()
```
